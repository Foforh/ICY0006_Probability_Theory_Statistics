```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams

rcParams["figure.figsize"] = (10, 6)
sns.set()
```

```python
main_data = pd.read_csv("FuelConsumptionCo2.csv")
```

```python
main_data.sample(5)
```

###### Observation: Here we can see most of the car have M6, AS6 or A6 Transmission
